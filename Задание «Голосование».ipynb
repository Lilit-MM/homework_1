```python
citizens_votes = [1, 2, 2, 3, 2, 1, 1, 3, 2, 3, 2, 2, 3, 3, 1, 2, 1, 3, 3, 1]
```

```python
votes_count = {}
for vote in (citizens_votes):
    if vote == 1:
        a = [1 for num in citizens_votes if num == 1]
        votes_count.setdefault('1', len(a))
    elif vote == 2:
        a = [1 for num in citizens_votes if num == 2]
        votes_count.setdefault('2', len(a))
    elif vote == 3:
        a = [1 for num in citizens_votes if num == 3]
        votes_count.setdefault('3', len(a))

print(votes_count)
```

```python
print(votes_count.keys())
print(votes_count.values())
votes_count.items()
```

```python
max_votes = max(votes_count.values())
```

```python
winners = []
for key, value in votes_count.items():
    if value == max_votes:
        winners.append(key)

print(winners)
```

```python
if len(winners) > 1:
    print("Нужен второй тур выборов")
else:
    print(winners)
```
